```python
import requests
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import nltk
```

```python
# Load processor and model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large")
```

```python
# Load the uploaded image from Colab
uploaded_image_path = "/content/pf-2.jpeg"  # Update with the actual uploaded image name
raw_image = Image.open(uploaded_image_path).convert('RGB')

# Conditional image captioning
inputs = processor(raw_image, return_tensors="pt")

out = model.generate(**inputs)
print(processor.decode(out[0], skip_special_tokens=True))

# Unconditional image captioning
inputs = processor(raw_image, return_tensors="pt")

generated_text = processor.decode(out[0], skip_special_tokens=True)
print("Generated Text (Conditional):", generated_text)
```

```python
from nltk import pos_tag, word_tokenize
from nltk.corpus import stopwords

# Tokenize and perform part-of-speech tagging
tokens = word_tokenize(generated_text)
pos_tags = pos_tag(tokens)

# Get English stop words from NLTK
stop_words = set(stopwords.words('english'))

# Extract nouns as potential keywords, excluding stop words
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("filtered tags:", filtered_tokens)
```

```python
categories_keywords = {
    'Refugee camps': ['shelter', 'refugee', 'displaced', 'tents'],
    'Floods': ['water', 'flooded'],
    'Fire': ['fire', 'burning', 'wildfire'],
    'Earthquake': ['building', 'demolished', 'rubble', 'destroyed'],
    'Police': ['police']
}

matching_category = []

for category, keywords in categories_keywords.items():
    for keyword in keywords:
        if keyword in filtered_tokens:
            matching_category.append(category)
        else:
            continue

if matching_category:
    print("Matching category:", matching_category)
else:
    print("No matching category found.")
```
